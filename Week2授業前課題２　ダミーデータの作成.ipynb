【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
[1.0 0.8
0.8 1.0]
多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。
numpy.random.multivariate_normal — NumPy v1.16 Manual
作成したndarrayのshapeは(500, 2)になります。
《シード値について》
実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておきましょう。今回シード値は0を使うことにします。

```python
import numpy as np
np.random.seed(0)
```

```python
mean1 = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
random_norm1 = np.random.multivariate_normal(mean1, cov, size = 500) 
```

```python
print(random_norm1)
```

【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。

散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
import matplotlib.pyplot as plt
x1_1 = random_norm1[:,0]
x2_1 = random_norm1[:,1]
plt.scatter(x1_1, x2_1)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.show()
```

【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
[1.0 0.8
0.8 1.0]
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
mean2 = [0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]]
random_norm2 = np.random.multivariate_normal(mean2, cov, size = 500) 
```

```python
x1_2 = random_norm2[:, 0]
x2_2 = random_norm2[:, 1]
plt.scatter(x1_1, x2_1, label="0")
plt.scatter(x1_2, x2_2, label="1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.legend(loc='upper right')
plt.show()
```
