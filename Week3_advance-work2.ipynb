```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
miss_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
miss_data[miss_data['Total']>0]
```

③　5つ以上欠損値がある特徴量（列）は削除する。

```python
df_drop = df.drop((miss_data[miss_data['Total']>5]).index, 1)
```

```python
df_drop
```

<b>④　欠損値があるサンプル（行）は削除する。<b>

```python
df_drop = df_drop.drop(df_drop.loc[df_drop["Electrical"].isnull()].index)
```

 # 【問題6】分布の確認
①seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

②目的変数に対して対数変換を行う。

③対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

④対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

①seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

```python
import matplotlib.pyplot as plt
import seaborn as sns
df_saleprice = df_drop["SalePrice"]
sns.distplot(df_saleprice, bins=100)
plt.show()
```

```python
print("歪度： %.2f" % df_saleprice.skew())
print("尖度：　%.2f" % df_saleprice.kurt())
```

②目的変数に対して対数変換を行う。

```python
import numpy as np
df_sp_log = np.log(df_saleprice)
df_sp_log
```

③対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

```python
sns.distplot(df_sp_log, bins=100)
plt.show()
print("歪度： %.2f" % df_sp_log.skew())
print("尖度：　%.2f" % df_sp_log.kurt())
```

④対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

対数化したことで正規分布にかなり近づいた。

《補足》

重回帰分析などで予測する場合、相関が高すぎる特徴量の組み合わせがあると予測に悪影響が出ることがあります。このような場合の簡単な対処方法として、そういった特徴量の組みを探し出して、１つに絞るということがあります。

①相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

```python
hm = df_drop.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(hm)
```

②目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

```python
hml = hm.nlargest(11, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_drop[hml].values.T)
sns.heatmap(cm, yticklabels=hml.values, xticklabels=hml.values, fmt=".2f", annot=True, vmax=1.0, vmin=0)

plt.tight_layout()
plt.show()
```

③選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。

・OverallQual　・・・全体的な品質
